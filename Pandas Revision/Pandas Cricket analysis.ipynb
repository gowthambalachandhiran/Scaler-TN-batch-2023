```python
df['Innings Boundary Sixes'].unique()
```

```python
def replace_spl_chr(field):
    if field == '-':
        return '0'
    else:
        return field
```

```python
df['Innings Boundary Sixes'] = df['Innings Boundary Sixes'].apply(replace_spl_chr)
```

```python
df['Innings Boundary Sixes'] = df['Innings Boundary Sixes'].astype(float)
```

```python
df['Innings Boundary Sixes'].unique()
```

```python
df.info()
```

```python
def not_out_ind(field):
    if '*' in field:
        return 1
    else:
        return 0
```

```python
df['Innings Runs Scored'].unique()
```

```python
df['Not out'] = df['Innings Runs Scored'].apply(not_out_ind)
```

```python
df['Innings Boundary Sixes'].max()
```
