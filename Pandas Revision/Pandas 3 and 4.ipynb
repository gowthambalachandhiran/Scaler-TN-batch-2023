# Pandas Melt

```python
df = df[['Player','Span','Mat']]
```

```python
df.head()
```

```python
df_melted = pd.melt(df,id_vars=["Player"],value_vars=['Span','Mat'],var_name='Attribute',value_name='value')
```

# Pivot operation
