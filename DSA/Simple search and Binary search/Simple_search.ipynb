##Can we do binary search and optimize the current setup

Algorithm
- Choose a middle element
 - left = 0
 - right = len(B)
 - middle element = left+ right//2
 - if target == middle return middle
 - if target > middle then new left = middle + 1
 - if target < middle then new right = middle - 1

 - calculate the new middle = left + right // 2
 - after checking the above condition return target not found



```python
B = [1,8,13,17,1000]
target = 1000
left = 0
right = len(B) - 1
while left <= right:
  middle = (left+right)//2
  if B[middle] == target:
    print(target)
    break
  elif target > B[middle]:
    left = middle + 1
  else:
    right = middle - 1
print(-1)
```
