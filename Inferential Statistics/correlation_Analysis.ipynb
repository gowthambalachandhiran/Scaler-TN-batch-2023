```python
plant_heights = np.array([15, 20, 25, 18, 22, 30, 28, 35])

sunlight_rankings = np.array([4, 7, 3, 6, 5, 8, 2, 1])
```

```python
# Scatter plot
plt.scatter(plant_heights,sunlight_rankings,cmap='hot')

# Display the plot
plt.show()
```

```python
# H0: No correlation
# Ha: There is correlation
spearmanr(plant_heights,sunlight_rankings)
```
