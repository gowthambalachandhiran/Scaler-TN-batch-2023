You are a product manager for a company that has recently launched a new product.
Customer satisfaction is a critical metric, and you want to determine if the proportion of satisfied customers with the new product meets your target satisfaction level of 70%.
You collected a random sample of 150 customer reviews, and 115 of them expressed satisfaction with the product.Screenshot 2024-03-08 at 10.57.03â€¯PM.png

```python
import numpy as np
import scipy.stats as stats
```

```python
satisfied_customers = 115
total_customers = 150
target_satisfaction = 0.70
```

- Ho:  p = p_hat
- Ha: p < p_hat

```python
p = target_satisfaction
```

```python
p_hat = satisfied_customers / total_customers
```

```python
p_hat
```

```python
n = total_customers
```

```python
z_st = (p_hat - p) / np.sqrt((p * (1 - p)) / n)
```

```python
print(z_st)
```

```python
p_value = 2 * ( 1- stats.norm.cdf(np.abs(z_st)))
print(p_value)
```

```python
# Print the results
print(f"Z-statistic = {z_st}")
print(f"P-value = {p_value}")
```

```python
alpha = 0.05  # Significance level
if p_value < alpha:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

if decision == "Reject the null hypothesis":
    conclusion = "There is enough evidence to conclude that the proportion of satisfied customers with the new product is statistically different from the target satisfaction level."
else:
    conclusion = "There is not enough evidence to conclude that the proportion of satisfied customers with the new product is statistically different from the target satisfaction level."

# Print the results
print(f"Decision: {decision}")
print(f"Conclusion: {conclusion}")
```

A researcher is studying the satisfaction level of customers after implementing a new customer service system. They collected survey responses from 250 customers and found that 65 of them were dissatisfied with the new system.
The researcher wants to test the null hypothesis that no more than 30% of customers are dissatisfied with the new system.
Use the p-value technique to test the claim with a significance level of Î± = 0.05.

```python
n = 250
x = 65
p_hat = x/n
p = 0.30
```

```python
z_st = (p_hat - p) / np.sqrt((p * (1 - p)) / n)

print(z_st)
```

```python
p_value = 1 - stats.norm.cdf(z_st)
```

```python
print(p_value)
```

```python
# Define the significance level
alpha = 0.05

# Make a decision based on the p-value and significance level
if p_value < alpha:
  print('Reject the null hypothesis.')
else:
  print('Fail to reject the null hypothesis.')
```

You are the manager of an e-commerce website, and you have recently implemented a new web page in hopes of increasing sales.

To evaluate the effectiveness of the new page, you collected data on the conversion rates for both the old and new web pages. The conversion rate is defined as the proportion of visitors who make a purchase.

For the old web page (Web Page A), you had 1000 visitors, resulting in 50 conversions.
For the new web page (Web Page B), you had 500 visitors, resulting in 30 conversions.
Now, you want to determine if there is a statistically significant difference in the conversion rates between the old and new web pages.

```python
import statsmodels.api as sm
import numpy as np
```

```python
# Define the data
conversions = np.array([50, 30])  # Number of conversions for Web Page A and Web Page B # x_1, x_2
visits = np.array([1000, 500])
```

```python
conversions
```

```python
visits
```

```python
z_stat, p_value = sm.stats.proportions_ztest(conversions,
                           visits,
                           alternative='greater')
```

```python
# Print the results
print(f"Z-statistic = {z_stat}")
print(f"P-value = {p_value}")
```

```python
# Data for the old web page (Web Page A)
n_1 = 1000 # visits_1
x_1 = 50 # conversions_1

# Data for the new web page (Web Page B)
n_2 = 500 # visits_2
x_2 = 30 # conversions_2
```

```python
p_1_hat = x_1 / n_1
p_2_hat = x_2 / n_2
```

```python
p_hat = (x_1 + x_2) / (n_1 + n_2)
```

```python
Z = (p_1_hat - p_2_hat) / np.sqrt(p_hat * (1 - p_hat) * ((1 / n_1) + (1 / n_2)))
```

```python
p_value = 2 * (1 - stats.norm.cdf(np.abs(Z)))
```

```python
# Make a decision and conclusion
if p_value < alpha:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

if decision == "Reject the null hypothesis":
    conclusion = "There is enough evidence to conclude that the conversion rates of the new web page (Web Page B) are statistically different from the old web page (Web Page A)."
else:
    conclusion = "There is not enough evidence to conclude that the conversion rates of the new web page (Web Page B) are statistically different from the old web page (Web Page A)."

# Print the results
print(f"Decision: {decision}")
print(f"Conclusion: {conclusion}")
```

## Option1: Ha = u1 != u2

```python
from scipy.stats import ttest_ind
```

```python
iq_1 = df_iq[df_iq['School'] == 'school_1']['iq']
iq_2 = df_iq[df_iq['School'] == 'school_2']['iq']

iq_1
```

```python
t_stat, pvalue = ttest_ind(iq_1, iq_2)
t_stat, pvalue
```

```python
alpha = 0.05 # 95% confidence

if pvalue < alpha:
  print('Reject H0')
else:
  print ('Fail to Reject H0')
```

## Option2: Ha => u1 > u2

```python
t_stat, pvalue = ttest_ind(iq_1, iq_2, alternative = "greater")
t_stat, pvalue
```

```python
alpha = 0.05 # 95% confidence

if pvalue < alpha:
  print('Reject H0')
else:
  print ('Fail to Reject H0')
```

## ## Option3: Ha => u1 < u2

```python
t_stat, pvalue = ttest_ind(iq_1, iq_2, alternative = "less")
t_stat, pvalue
```

```python
alpha = 0.05 # 95% confidence

if pvalue < alpha:
  print('Reject H0')
else:
  print ('Fail to Reject H0')
```

```python
Path= '/content/problem_solving.csv'
df_ps = pd.read_csv(Path)
```

```python
# H0: Problem-solving has no effect
# Ha: Problem-solving had an effect
```

```python
statistic, pvalue = ttest_rel(df_ps["test_1"], df_ps["test_2"])
print("Test statistic:",statistic)
print("pvalue:",pvalue)
```

```python
alpha = 0.05 # 95% confidence

if pvalue < alpha:
  print('Reject H0')
else:
  print ('Fail to Reject H0')
```

```python
# H0: Problem solving has no effect
# Ha: Problem solving improved the scores

statistic, pvalue = ttest_rel(df_ps["test_1"], df_ps["test_2"],alternative="less")
print("Test statistic:",statistic)
print("pvalue:",pvalue)
```

```python
alpha = 0.05 # 95% confidence

if pvalue < alpha:
  print('Reject H0')
else:
  print ('Fail to Reject H0')
```
